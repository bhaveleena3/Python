```python
import pandas as pd
from sklearn.svm import SVC
import numpy as np
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
from sklearn.preprocessing import LabelEncoder
```

```python
Salary=pd.read_csv('/content/SalaryData_Test(1).csv')
Salary
```

```python
Salary1=pd.read_csv('/content/SalaryData_Train(1).csv')
Salary1
```

```python
List=Salary1.append(Salary)# append two files and make 1
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder() #encoding column
List['workclass'] = label_encoder.fit_transform(List['workclass'])
List['education'] = label_encoder.fit_transform(List['education'] )
List['maritalstatus'] = label_encoder.fit_transform(List['maritalstatus'])
List['occupation'] = label_encoder.fit_transform(List['occupation'])
List['relationship'] = label_encoder.fit_transform(List['relationship'])
List['race'] = label_encoder.fit_transform(List['race'])
List['sex'] = label_encoder.fit_transform(List['sex'])
List['native'] = label_encoder.fit_transform(List['native'])
```

```python
List
```

```python

array=List.values
X = array[:,0:-1]
Y = array[:,-1]
```

```python
Xtest,Xtrain,Ytest,Ytrain=train_test_split(X,Y,test_size=0.3)
```

```python
clf = SVC(kernel='rbf',gamma=0.0001)
clf.fit(Xtrain,Ytrain)
```

```python
Pred=clf.predict(Xtest)
Pred
```

```python
accuracy_score(Pred,Ytest)
```

```python
confusion_matrix(Pred,Ytest)
```
