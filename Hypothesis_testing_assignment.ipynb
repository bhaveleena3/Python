Example 1

H1:There is  significant diffrence in diameter of cutlet between two units.
H0: There is no significant diffrence in diameter of cutlet between two units.

```python
import pandas as pd
from scipy import stats

```

```python
df=pd.read_csv('/content/Cutlets.csv')
```

```python
df
```

```python
stats.t.ppf(0.975,68)
```

```python
stats.ttest_ind(df['Unit A'],df['Unit B']) # two tail independent
```

p value (0.47) > alpha (0.05) so H1 is accepted.
There is diffrence in diameter of cutlet between two units.
