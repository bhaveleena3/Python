Build a prediction model for Salary_hike


```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LogisticRegression
import statsmodels.formula.api as smf
```

```python
#read data
df=pd.read_csv('/content/Salary_Data (1).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#check for missing values
df.isna().sum()
```

```python
df.corr()
```

Scatterplot between variables along with histograms
Check for Assumption: Independent variables are independent of each other

```python
sns.pairplot(df)
```

```python
SLModel=smf.ols('YearsExperience~Salary',data = df).fit() #Simple Linear Regression model for YearsExperience Vs Salary
```

```python
print(SLModel.tvalues, '\n', SLModel.pvalues)
```

```python
SLModel.summary() #95% acuracy and P value=0
```

B0=-2.38,B1=0.0001

```python
pre=SLModel.predict(df["Salary"])
SLModel.resid #error
```

```python
sns.regplot(x="Salary",y="YearsExperience",data=df) #both having positive correlation and Best fit line
```
